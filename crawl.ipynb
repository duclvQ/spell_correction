```python
json_path = "freq_list_results.json"
```

```python
import json 
data = json.load(open(json_path))
data = data['data']
```

```python
data
```

```python
word_list = list()
with open('vi_freq_dict.txt', 'w', encoding='utf-8') as f:
    for d in data:
        word = d['word']
        occurrences = d['count']
        # add_underscore
        word = word.replace(" ", "_")
        f.write(f"{word}\t{occurrences}\n")
```
